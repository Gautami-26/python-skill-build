## Class

Class is very imp in python, class in Python is like a blueprint used to create objects. It groups related data (variables) and functions (methods) together so you can use them easily.

Like instagram have so much features in it

```python
class Dog:
    def __init__(self, name, age):
        self.name=name
        self.age=age

    def bark(self):
        return f'{self.name} says Bhoww.. bhow...!!'

my_dog=Dog("Marshall",4)

print(my_dog.bark())

```

```python
class Student:
    def __init__(self, name, age, marks):
        self.name=name
        self.age=age
        self.marks=marks

    def show_student_data(self):
        print(f'Hi my name is {self.name} and my age is {self.age} and my marks are {self.marks}')



s1=Student("Gautami",19,99)
s2=Student("Mayuri",12,95)
s3=Student("Sanika",21,95)

s1.show_student_data()
s2.show_student_data()
s3.show_student_data()
```

```python

class Library:
    def __init__(self, name,book):
        self.name=name
        self.book=book
    
    def show_data(self):
        print(f'Author name is {self.name} and their book name is {self.book}.')



s1=Library("Gautami","Data Science")
s2=Library("Gautami","Physics")
s3=Library("Sanika","Chemistry")
s4=Library("Gauri","Maths")
s5=Library("Sakshi","Geometry")

s1.show_data()
s2.show_data()
s3.show_data()
s4.show_data()
s5.show_data()
```

```python
class Circle:
    #circle_aea=pi*r*r
    # perimeter=2*pi*r
    def __init__(self,pi,r):
        self.pi=pi
        self.r=r
    def area(self):
        return self.pi*self.r*self.r

    def perimeter(self):
        return 2*self.pi*self.r
s1=Circle(3.14,4)

result1=s1.area()
result2=s1.perimeter()

print("Area of Circle:",result1)
print("Perimeter of Circle:",result2)
```

```python
class List_item:
    def __init__(self):
        self.items = []

    def add_items(self, item):
        self.items.append(item)

    def remove_items(self, item):
        self.items.remove(item)

    def view_all(self):
        print(self.items)


list_call = List_item()

# Just calling functions one by one
list_call.add_items("apple")
list_call.add_items("banana")
list_call.view_all()        # shows items

list_call.remove_items("apple")
list_call.view_all()        # shows items after removing
```

```python
class List_item:
    def __init__(self):
        self.items = []

    def add_items(self, item):
        self.items.append(item)

    def remove_items(self, item):
        if item in self.items:
            self.items.remove(item)
        else:
            print("Item not found!")

    def view_all(self):
        print("Current items:", self.items)


list_call = List_item()

while True:
    print("\nEnter your choice:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View all items")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        item = input("Enter item to add: ")
        list_call.add_items(item)

    elif choice == 2:
        item = input("Enter item to remove: ")
        list_call.remove_items(item)

    elif choice == 3:
        list_call.view_all()

    elif choice == 4:
        print("Exiting the program...")
        break

    else:
        print("Invalid choice!")
```
