```python
import time
def timer(func):
    def wrapper():
        start=time.time()
        func()
        end=time.time()
        print("Time:",end-start)
    return wrapper

@timer
def process():
    for i in range(1000000000):
        pass

process()        
```

```python
def logging_funda(func):
    def wrapper(*a,**k):
        print("Function name:",func.__name__)
        print("Postional args(*a):",a)
        print("Keyword args (**k):",k)
        return func(*a,**k)
    return wrapper

@logging_funda
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)

@logging_funda
def fibonaaci(n):
    if n<=1:
        return n
    return fibonacci(n-1)+fibonaaci(n-2)

print("factorial result:",factorial(5))
print()
print("Fibonacci result:",fibonacci(6))
```
