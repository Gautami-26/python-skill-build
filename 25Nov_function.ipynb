```python
# Prime Number

def is_prime(n):
    if n <= 1:
        return f"{n} is not a prime number"
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return f"{n} is not a prime number"
    return f"{n} is a prime number"

print(is_prime(17))
```

```python
# Fibonacci Series

def fibonacci(n):
    if n <= 0:
        return "Enter a positive number"
    
    a = 0
    b = 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    
    return series

result = fibonacci(10)
print(result)

```

```python
# Factorial

def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return f"Factorial of {n} is {result}"
print(factorial(5))
```
