```python
# find even number

def even_num(a):
    if a%2==0:
        return f"{a} is even"
    
    else:
        return f"{a} is not even"
    
result=even_num(17)    

print(result)
```

```python
def list_sum(list_of_numbers):
    return sum(list_of_numbers)

list_of_numbers=[10,20,30]
result=list_sum(list_of_numbers)

print(result)
```

```python
def reverse_string(the_string):
    return the_string[::-1]

the_string="gautami"
result=reverse_string(the_string)

print(result)
```

```python
def is_palindrome(text):
    s=str(text).lower()
    return s== s[::-1]

print(is_palindrome('nitin'))
```
