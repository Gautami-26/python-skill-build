```python
def div_num():

    try:
        a=12
        b=4

        print("Result:",a/b)

    except (ZeroDivisionError, ValueError):
        print("some error, please check it thoroughly....")
    finally:
        print("terminating from here")   
div_num()        
```

```python
def factorial(n):
    try:
        if n<0:
            raise ValueError("caanot be a negative number")
        if n==0:
            return 1
        return n*factorial(n-1)
    except ValueError as e:
        return e
print(factorial(-5))    
```
