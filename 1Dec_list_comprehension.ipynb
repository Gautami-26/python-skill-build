## do dictionary in list comprehension

```python
my_dict = {"name": "Gautami", "age": 21, "city": "Pune"}

keys = [k for k in my_dict]
print(keys)
```

```python
values = [v for v in my_dict.values()]
print(values)
```

```python
pairs = [(k, v) for k, v in my_dict.items()]
print(pairs)
```

```python
only_int_values = [v for v in my_dict.values() if type(v) == int]
print(only_int_values)
```

```python
key_lengths = [len(k) for k in my_dict]
print(key_lengths)
```

```python
formatted = [f"{k}: {v}" for k, v in my_dict.items()]
print(formatted)
```

```python
product = {"pen": 10, "book": 50, "bag": 300}

new_prices = {k: v+10 for k, v in product.items()}
print(new_prices)
```

```python
filtered = {k: v for k, v in product.items() if v > 50}
print(filtered)
```
