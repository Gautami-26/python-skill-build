```python
def linear_search(inventory, target_id):
    for item in inventory:
        if item['id'] == target_id:
            return item
    return None

print("Linear Search Result:", linear_search(blinkit_inventory, 1002))
```

```python
def binary_search(items, target_id):
    items = sorted(items, key=lambda x: x['id']) 

    low = 0
    high = len(items) - 1

    while low <= high:
        mid = (low + high) // 2
        if items[mid]['id'] == target_id:
            return items[mid]
        elif target_id < items[mid]['id']:
            high = mid - 1
        else:
            low = mid + 1

    return "Item not found"

print("Binary Search Result:", binary_search(blinkit_inventory, 1002))
```
