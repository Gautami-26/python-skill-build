```python
import sqlite3

conn=sqlite3.connect("school.db")
cursor =conn.cursor()

cursor.execute("""
CREATE TABLE Students(
               id INTEGER,
               name TEXT,
               marks INTEGER
)               
""")

conn.commit()
conn.close()
```

```python
import sqlite3

conn=sqlite3.connect("school.db")
cursor =conn.cursor()

cursor.execute("INSERT INTO students VALUES (1,'Rahul',85)")

conn.commit()
conn.close()
```
