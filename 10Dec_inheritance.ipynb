## Inheritance

```python
class A:
    def fromClassA(self):
        print("This is from class A")

class B(A):
    def fromClassB(self):
        print("This is from class B")

obj=B()
obj.fromClassA()
obj.fromClassB()        
            
```

```python
class Parent:
    def greet(self):
        print("Hello from parent class")

class Child(Parent):
    pass

c=Child()
c.greet()
```

```python
#  Multiple Inheritance
class A:
    def fromClassA(self):
        print("This is from class A")

class B(A):
    def fromClassB(self):
        print("This is from class B")

class C(A, B):
    pass


objC=C()
objB.fromClassA()
objC.fromClassB() 
 
```

```python
#  Multiplr Inheritance
class A:
    def show(self):
        print("This is from class A")

class B(A):
    def show(self):
        print("This is from class B")

class C(A):
    def show(self):
        print("This is from Class C")

class D(B, C):
    pass


objD=D()
objD.show()
 
```

## Polymorphism

```python
class Ethans:
    def DataScienceCourse(self):
        print("Stats, Maths, ML ,DL, NLP etc in detail")

class Others:
    def DataScienceCourse(self):
        print("Overview with only programming ,not much in detail")        

for inst in (Ethans(), Others()):
    inst.DataScienceCourse()
```

```python
class Parent:
    def show(self):
        print("This is from parent")

class Child:
    def show(self):
        print("This is from child")

Child().show()                
```

```python
class Payment:
    def pay(self):
        print("This is from Payment class")

class Card(Payment):
    def pay(self):
        print("Card Payment Done!!")

class UPI(Payment):
    def pay(self):
        print("UPI payment done")

for p  in [Card(), UPI()]:
    p.pay()
```
