```python
def greet(func):
    def wrapper():
        print("Good Morning sir, may i come in")
        func()
    return wrapper

@greet
def attendance_register():
    print("I wnt this attendance register")

    attendance_register()    
```

```python
def greet_stud(func):
    def wrapper():
        print("Before we call func() wala function")
        print("hello shivangi")
        func()

        print("func to cll karne k baad vali line")
    return wrapper

@greet_stud
def call_stud_name():
    print("Shiv ne swaal pucha")

call_stud_name()    

```

```python

class AWSPipeline:
    def load_config(self):
        print("Loading Configuration")
    def connect_db(self):
        print("Connceting to db")

class SalesPipeline(AWSPipeline):
    def run(self):
        self.load_config()
        self.connect_db()
        print("Running sales data pipeline") 

class MarketingPipeline(AWSPipeline):
    def run(self):
        self.load_config()
        self.connect_db()
        print("Running marketing data")                   
```
