```python
dictionary_1={
    "Gautami":"Data Scientist",
    "Mayuri":"UI UX developer"
}

dictionary_2={
    "Sahil":"Python developer",
    "Gauri":"Team Leader",
    "Sakshi":"Frontend Developer"
}
```

```python
merged_dict={**dictionary_1,**dictionary_2}
```

```python
merged_dict
```

```python
for key1, value1 in merged_dict.items():
    print(key1,":",value1)
```

```python
for key1, value1 in merged_dict.items():
    print(key1,value1)
```
