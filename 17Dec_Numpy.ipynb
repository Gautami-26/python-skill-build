```python
A=np.matrix([[1,2,3,33],
            [4,5,6,66],
            [7,8,9,99]])
np.argmax(A)  #11,which is the position of 99
```

```python
np.argmax(A[:,:]) #11,which is the position of 99
```

```python
np.argmax(A[:1])    #3, which is the position of 33
```

```python
np.argmax(A[:,2])  #2,which is the position of 9
```
