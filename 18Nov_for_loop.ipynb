```python
num=int(input("Enter the number"))

if num<2:
    print("it is coprime number")
else:
    is_prime_flag = True
    for i in range(2, num//2):
        print("Current denominator i.e. i: ",i) 
        print(f"Checking if {num}//{i} gives 0 remainder")
        if num%i==0:
            print(f"Wow!! {i} is 1 of the divisor")
            is_prime_flag = False
            break
    if is_prime_flag:
        print("The number entered is Prime") 
    else:
        print("Its a non prime number")       
```
